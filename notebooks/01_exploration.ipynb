```python
# 1. Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# 2. Load Dataset
df = pd.read_csv("D:\Intership\sales_forecasting_dashboard\Data\Raw\Superstore.csv", encoding='ISO-8859-1') 
df.head()
```

```python
# 3. Overview of Data
print("Shape:", df.shape)
df.info()
df.describe(include='all')
```

```python
# 4. Convert Order Date and Ship Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Confirm conversion
df[['Order Date', 'Ship Date']].dtypes
```

```python
# 5. Basic Time Range Information
print("Earliest Order Date:", df['Order Date'].min())
print("Latest Order Date:", df['Order Date'].max())
```

```python
# 6. Sales Over Time (Raw Daily Plot)
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()

import seaborn as sns
sns.boxplot(x=daily_sales['Sales'])
plt.title("Boxplot of Daily Sales")
plt.show()
```

```python
Q1 = daily_sales['Sales'].quantile(0.25)
Q3 = daily_sales['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = daily_sales[(daily_sales['Sales'] < lower_bound) | (daily_sales['Sales'] > upper_bound)]
print("Number of outlier days:", len(outliers))
outliers.head()
```

```python
# Cap the sales values
daily_sales['Sales_capped'] = np.where(
    daily_sales['Sales'] > upper_bound, upper_bound, daily_sales['Sales']
)

plt.figure(figsize=(12, 5))
plt.plot(daily_sales['Order Date'], daily_sales['Sales'], label='Original')
plt.plot(daily_sales['Order Date'], daily_sales['Sales_capped'], label='Capped', alpha=0.7)
plt.title("Original vs Capped Sales")
plt.legend()
plt.show()
```

```python
# 7. Monthly Sales Trend (Prophet Ready)
monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum().reset_index()
monthly_sales.columns = ['ds', 'y']  # Prophet format

plt.figure(figsize=(14, 5))
plt.plot(monthly_sales['ds'], monthly_sales['y'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# 8. Regional Breakdown

df['Region'].unique()

# Monthly sales per region
region_monthly_sales = (
    df.groupby([pd.Grouper(key='Order Date', freq='M'), 'Region'])['Sales']
    .sum()
    .reset_index()
)

# Plot region-wise trends
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.lineplot(data=region_monthly_sales, x='Order Date', y='Sales', hue='Region')
plt.title("Monthly Sales by Region")
plt.show()
```

```python
 # 9. Export Prophet-Ready CSV Files

# National-level monthly sales
national_monthly = daily_sales.resample('M', on='Order Date')['Sales_capped'].sum().reset_index()
national_monthly.columns = ['ds', 'y']
national_monthly.to_csv('D:\Intership\sales_forecasting_dashboard\Data\Processed\monthly_sales_national.csv', index=False)

# Region-wise monthly sales
region_monthly = df.groupby([pd.Grouper(key='Order Date', freq='M'), 'Region'])['Sales'].sum().reset_index()
region_monthly.columns = ['ds', 'Region', 'y']
region_monthly.to_csv('D:\Intership\sales_forecasting_dashboard\Data\Processed\monthly_sales_by_region.csv', index=False)
```
